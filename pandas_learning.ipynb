```python
# Imports and setup
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_theme(style="whitegrid")
```

```python
# Load dataset
df = pd.read_csv("datasets/organizations.csv", index_col=0)
```

```python
# Inspect datatset
df.describe(include="all")
```

```python
df.head(10)
```

```python
# Clean data
df = df.drop_duplicates()
df.head(10)
```

```python
# Top industries
plt.figure(figsize=(8, 4))
top10 = df["Industry"].value_counts().index[:10]
sns.countplot(y="Industry", data=df[df["Industry"].isin(top10)], order=top10)
plt.title("Top 10 Industries by Organization Count")
plt.xlabel("Count")
plt.ylabel("Industry")
plt.show()
```

```python
plt.figure(figsize=(8, 4))
sns.histplot(df["Founded"], bins=20, kde=True)
plt.title("Distribution of Founded Years")
plt.xlabel("Year Founded")
plt.ylabel("Number of Organizations")
plt.show()
```
